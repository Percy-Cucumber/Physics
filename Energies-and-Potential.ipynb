```python
%matplotlib notebook

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

k = 1

def energy(q, r, rhat):
    k*(q/r**2)*rhat
    
t = -100
tl = []
dt = 0.001

q1 = -5
q2 = +2
q3 = +2

q1p = -5
q2p = 0
q3p = 5

E1l = []
E2l = []
E3l = []
ETl = []

dV1 = []
dV2 = []
dV3 = []
dVT = []

Z1 = []
Z2 = []

while t < 100:
    E1 = k*(q1/(t - q1p)**2)*((t - q1p)/abs(t - q1p))
    E2 = k*(q2/(t - q2p)**2)*((t - q2p)/abs(t - q2p))
    E3 = k*(q3/(t - q3p)**2)*((t - q3p)/abs(t - q3p))
    ET = E1+E2+E3
    
    V1 = E1*(t - q1p)
    V2 = E2*(t - q2p)
    V3 = E3*(t - q3p)
    VT = V1+V2+V3
    
    
    Z1.append()
    
    E1l.append(E1)
    E2l.append(E2)
    E3l.append(E3)
    ETl.append(ET)
    
    dV1.append(V1)
    dV2.append(V2)
    dV3.append(V3)
    dVT.append(VT)
    
    tl.append(t)
    t += dt


fig1, ax1 = plt.subplots()
plt.plot(tl, E1l, 'o', markersize=0.05, label='Q1 (blue)')
plt.plot(tl, E2l, 'o', markersize=0.05, label='Q2 (orange)')
plt.plot(tl, E3l, 'o', markersize=0.05, label='Q3 (green)')
ax1.set(xlabel='distance (m)', ylabel='Energy (N/C)', title='Energy at charged particles')
plt.ylim([-10,10])
plt.axis([-100,100,-10,10])
plt.legend()

fig2, ax2 = plt.subplots()
plt.plot(tl, ETl, 'o', markersize=0.05, label='Total Energy (V/m)')
ax2.set(xlabel='distance (m)', ylabel='Energy (N/C)', title='Energy at charged particles')
plt.legend()
plt.ylim([-10,10])
plt.axis([-100,100,-10,10])
plt.grid()

fig3, ax3 = plt.subplots()
plt.plot(tl, dV1, 'o', markersize=0.05, label='V1')
plt.plot(tl, dV2, 'o', markersize=0.05, label='V2')
plt.plot(tl, dV3, 'o', markersize=0.05, label='V3')
ax3.set(xlabel='distance (m)', ylabel='Change in potential', title='Potential of each charge')
plt.legend()
plt.ylim([-10,10])
plt.axis([-100,100,-10,10])

fig4, ax4 = plt.subplots()
plt.plot(tl, dVT, 'o', markersize=0.1, label='Total Potential')
ax4.set(xlabel='distance (m)', ylabel='Change in potential', title='Total Potential (V)')
plt.legend()
plt.ylim([-10,10])
plt.axis([-100,100,-10,10])
plt.grid()

plt.show()
```
