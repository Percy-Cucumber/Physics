```python
import csv
import vpython as vp

def trendline(i):
    return 2e-5*i+0.0019

t = 0
ti = 2811.222
index = 901

vp.scene.background = color=vp.color.white

csvfile = open('Stuffed.csv', 'r')
reader = csv.reader(csvfile, delimiter=',', quotechar='|')
listreader = list(reader)

# Objects
car = vp.box(pos=vp.vec(0,0,0), size=vp.vec(2,1,0.2), color=vp.vec(0.4,0.6,0), up=vp.vec(0,0,1), make_trail=True)
xarrow = vp.arrow(pos=car.pos, shaftwidth=0.3, length=2)
yarrow = vp.arrow(pos=car.pos, shaftwidth=0.3, length=2)
zarrow = vp.arrow(pos=car.pos, shaftwidth=0.3, length=2)

bar = vp.box(pos=vp.vec(-9.5,-1,0), size=vp.vec(19,0.4,0.4), color=vp.color.black)

gv = vp.gcurve(color=vp.vec(0,0.3,0.7))

yi = vp.vec(0,1,0)
xi = vp.vec(1,0,0)
zi = vp.vec(0,0,1)

yvi = 4

while index < 1436:
    vp.rate(20)

#     Timestamp /dt
    listfile = listreader[index]
    timelist = listfile[0]
    tf = float(timelist[0:2])*60 + float(timelist[3:5]) + float(timelist[6])*0.1
    dt = tf - ti
    
    t += dt

#     Acces acceleration
    accelX = (float(listfile[4]))*(9.81)
    accelY = (float(listfile[5])-trendline(index))*(9.81)
    accelZ = (float(listfile[6]))*(9.81)
    
    gyroX = float(listfile[7])*dt
    gyroY = float(listfile[8])*dt
    gyroZ = float(listfile[9])*dt
    
    yf = vp.rotate(vp.rotate(yi, gyroY, xi), gyroZ, zi)
    xf = vp.rotate(vp.rotate(xi, gyroX, yi), gyroZ, zi)
    zf = vp.rotate(vp.rotate(zi, gyroY, xi), gyroX, yi)
    
    yhat = vp.norm(yf)
    xhat = vp.norm(xf)
    zhat = vp.norm(zf)
    
    yv = yhat*yvi + yhat*accelY*dt
    xv = xhat*accelX*dt
    zv = zhat*accelZ*dt
    
    car.pos = car.pos + (yv+xv+zv)*dt
    yarrow.pos = car.pos
    xarrow.pos = car.pos
    zarrow.pos = car.pos
    
    car.axis = yhat
    yarrow.axis = yhat
    xarrow.axis = xhat
    zarrow.axis = zhat
    
    gv.plot(t, vp.mag(yv))

    vp.scene.caption = 'Time: ' + str(t)
    
    yi = yf
    xi = xf
    zi = zf
    
    yvi = vp.mag(yv)
    
    ti = tf
    index +=1

csvfile.close()
```
